```python
import numpy as np
import pandas as pd
# 1000 people
# 10 dollars each
# conduct a random coin flip: give the money to winner.
# repeat until only one player left
```

```python
# if random number > 0.5, second player wins
def update_players_with_outcome(players_dict, player_a, player_b):
    coin_flip = np.random.random()
    if coin_flip >= 0.5:
        players_dict[player_b] += 1
        players_dict[player_a] -= 1
    else:
        players_dict[player_b] -= 1
        players_dict[player_a] += 1
    return players_dict
```

```python
def save_state_for_plot(plot_dfs, players, count):
    plot_dfs.append(
        pd.DataFrame(
            {"player": players.keys(),
                "dollars": players.values(),
                "round": count}
                )
        )
```

```python
players = {i: 10 for i in range(100)}
count = 0
plot_dfs = []
still_playing = list(players.keys())


while len(still_playing) > 1 :
    if count % 10000 == 0:
        save_state_for_plot(plot_dfs, players, count)
    count +=1
    player_a, player_b = np.random.choice(still_playing, 2)
    players = update_players_with_outcome(players, player_a, player_b)
    for player_number in [player_a, player_b]:
        if players[player_number] == 0:
            still_playing.remove(player_number)


# add final state
save_state_for_plot(plot_dfs, players, count)
```

```python
plot_df = pd.concat(plot_dfs)
```

```python
import plotly.express as px
   
fig = px.bar(plot_df, 
             x ="player", 
             y ="dollars",
             animation_frame ='round',
             barmode="group",
             range_x=[0,100],
             range_y=[0,1000]
             )
fig.show()
```

```python
n_players_left = pd.DataFrame(
    plot_df[plot_df["dollars"] != 0]
    .groupby("round")
    ["player"].nunique()
    .reset_index()
    ).rename({"player": "n_players_left"}, axis=1)
```

```python
fig = px.line(n_players_left, 
             x ="round", 
             y ="n_players_left",
             )
fig.show()
```
